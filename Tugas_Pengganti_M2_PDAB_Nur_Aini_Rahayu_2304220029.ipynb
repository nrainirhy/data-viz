```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membaca file CSV dengan delimiter ";"
file_path = "/content/data_kendaraan.csv"  # Sesuaikan dengan lokasi file
df = pd.read_csv(file_path, delimiter=";", skipinitialspace=True)

# Menampilkan informasi dasar dataset
print(df.info())

# Menampilkan baris
print(df.head())
```

```python
# Set style
sns.set_style("whitegrid")

# Plot histogram
plt.figure(figsize=(8, 5))
sns.histplot(df["Konsumsi Bahan Bakar (km/l)"], bins=10, kde=True, color="skyblue")
plt.xticks(rotation=45)  # Memutar teks agar lebih terbaca
plt.show()

```

```python
# Pastikan kolom konsumsi bahan bakar dalam format numerik
df["Konsumsi Bahan Bakar (km/l)"] = df["Konsumsi Bahan Bakar (km/l)"].str.replace(',', '.').astype(float)

# Membuat scatter plot yang diperbaiki
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["Jarak Tempuh (km)"], y=df["Konsumsi Bahan Bakar (km/l)"], hue=df["Jenis Kendaraan"])
plt.title("Jarak Tempuh vs Konsumsi Bahan Bakar")
plt.xlabel("Jarak Tempuh (km)")
plt.ylabel("Konsumsi Bahan Bakar (km/l)")
plt.xticks(rotation=45)  # Memutar teks sumbu X jika diperlukan
plt.show()

```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x=df["Jenis Kendaraan"], palette="pastel")
plt.title("Jumlah Kendaraan Berdasarkan Jenis")
plt.xlabel("Jenis Kendaraan")
plt.ylabel("Jumlah")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
city_counts = df["Kota"].value_counts()
city_counts.plot(kind="bar", color="lightcoral")
plt.title("Sebaran Kendaraan per Kota")
plt.xlabel("Kota")
plt.ylabel("Jumlah Kendaraan")
plt.xticks(rotation=45)  # Rotasi nama kota agar lebih mudah dibaca
plt.show()
```
